```python
graph_df = pd.DataFrame()

# limits loop to 16 iterations
for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df['region'] == region]
    region_df.set_index("Date")
    region_df.sort_index(inplace=True)
    # Use f string to insert region names to keep them all from having the same index name in the next step
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        # Second set of brackets switches from slice to dataframe being return
        graph_df = region_df[[f'{region}_price25ma']]
        
    else:
        # Bring dataframes together on an index
        graph_df = graph_df.join(region_df[f'{region}_price25ma'])
```

```python
# fix ^ problem and start over
df = pd.read_csv("datasets/avocado.csv")
df = df.copy()[df['type'] == "organic"]
df['Date'] = pd.to_datetime(df["Date"])

# sort by columns instead of index
df.sort_values(by="Date", ascending=True, inplace=True)

graph_df = pd.DataFrame()

for region in df['region'].unique():
    print(region)
    region_df = df.copy()[df['region'] == region]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    # Use f string to insert region names to keep them all from having the same index name in the next step
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        # Second set of brackets switches from slice to dataframe being return
        graph_df = region_df[[f'{region}_price25ma']]
        
    else:
        # Bring dataframes together on an index
        graph_df = graph_df.join(region_df[f'{region}_price25ma'])
        
graph_df.tail()
```

```python
# make the graph bigger (8x5 inch)
# hide legend (too crazy)
# gap caused by NaN values
graph_df.plot(figsize=(8,5), legend=False)
```
